```python
 soup = BeautifulSoup(response.text, "html.parser")
```

```python
articles = soup.find_all("div", class_="news_content")
```

```python
news_list = []
```

```python
for article in articles:
        # ✅ Extract News Title
        title_tag = article.find("h2", class_="news_title")
        title = title_tag.get_text(strip=True) if title_tag else "No Title"

        # ✅ Extract News Description
        desc_tag = article.find("div", class_="news_desc")
        description = desc_tag.get_text(strip=True) if desc_tag else "No Description"

        # ✅ Append to list
        news_list.append({"Title": title, "Description": description})
```

```python
 df = pd.DataFrame(news_list)
```
