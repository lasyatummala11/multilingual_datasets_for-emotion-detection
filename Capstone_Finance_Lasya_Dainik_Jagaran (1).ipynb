```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from selenium import webdriver
```

```python
url = "https://www.bhaskar.com/business/"

response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
articles = soup.find_all("li", class_="c7ff6507 db9a2680")  # Adjust class names if needed
```

```python
news_list = []

for article in articles:
    title_tag = article.find("span", style="color:#3e9e3e")  # Headline selector
    desc_tag = article.find("div", class_="ad3ccf1a")  # Description selector
    link_tag = article.find("a")  # Extract link to the full article

    title = title_tag.get_text(strip=True) if title_tag else "No title found"
    description = desc_tag.get_text(strip=True) if desc_tag else "No description available"
    link = "https://your-news-website.com" + link_tag["href"] if link_tag and "href" in link_tag.attrs else "No link available"

    # Store the extracted data
    news_list.append({"Title": title, "Description": description, "Link": link})

# Print extracted news
for news in news_list:
    print(f"Title: {news['Title']}")
    print(f"Description: {news['Description']}")
    print(f"Link: {news['Link']}")
    print("-" * 100)
```

```python
df = pd.DataFrame(news_list)
df
```

```python
df.to_csv("business_news.csv", index=False, encoding="utf-8")
print("Scraping complete. Data saved to 'business_news.csv'")
```
