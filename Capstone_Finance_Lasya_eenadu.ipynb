```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from bs4 import BeautifulSoup
import requests
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import selenium
```

```python
response = requests.get(url)
print(response)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Extract all news articles
articles = soup.find_all("div", class_="thumb-description")

# List to store descriptions
descriptions = []

# Loop through each news article
for article in articles:
    # Extract description (within <br> tags)
    description_parts = []
    for br in article.find_all("br"):
        if br.next_sibling and isinstance(br.next_sibling, str):
            description_parts.append(br.next_sibling.strip())

    description_text = " ".join(description_parts)
    descriptions.append(description_text if description_text else "N/A")

# Create a DataFrame with only the 'Description' column
df = pd.DataFrame({"Description": descriptions})

# Display the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the target URL
url = "https://www.eenadu.net/business/latest-news"

# Send a request to fetch the HTML content
response = requests.get(url)
response.raise_for_status()  # Ensure request was successful

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract all news articles (links with headlines)
articles = soup.find_all("a", href=True)  # Find all <a> tags with href

# Lists to store data
headlines, links = [], []

# Loop through each article link
for article in articles:
    headline = article.text.strip()  # Get the text inside <a>
    link = article["href"]  # Extract the URL

    if headline and link.startswith("https://www.eenadu.net/telugu-news/business"):  # Filter relevant links
        headlines.append(headline)
        links.append(link)

# Create a DataFrame
df1 = pd.DataFrame({"Headline": headlines, "URL": links})

# Display DataFrame
print(df1)
```

```python
df1 = df1.drop(columns=['URL'])
df1
```

```python
df = df.rename(columns={'Description': 'Headlines'})
df
```

```python
df_combined = pd.concat([df, df1], ignore_index=True)

# Ensure the column name remains "Headlines"
df_combined.columns = ["Headlines"]
```
