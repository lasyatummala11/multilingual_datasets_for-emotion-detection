```python
import pandas as pd

# Load the dataset
file_path = "/cleaned_telugu_dataset.csv"
df = pd.read_csv(file_path)

# Pull 50 random rows
random_rows = df.sample(n=50, random_state=42)

# Display the sampled rows without using ace tools
random_rows
```

```python
# Define the output file path
output_file_path = "/random_telugu_sample.csv"

# Save the random sample to a CSV file
random_rows.to_csv(output_file_path, index=False)

# Provide the download link
output_file_path
```

```python
# Select only the 'Headline' and 'Content' columns and drop rows with null values
filtered_sample = random_rows[['Headline', 'Description']].dropna()

# Save the filtered sample to a CSV file
filtered_output_file_path = "/filtered_random_sample_telugu.csv"
filtered_sample.to_csv(filtered_output_file_path, index=False)

# Display the filtered sample
filtered_sample
```
