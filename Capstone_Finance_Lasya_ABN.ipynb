```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# ✅ Correct Base URL
base_url = "https://www.andhrajyothy.com/business/page/"

# ✅ Headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.90 Safari/537.36"
}

# ✅ List to store results
news_list = []

# ✅ Loop through pages 2 to 12
for page in range(2, 21):  # Pages 2 to 12
    url = f"{base_url}{page}"  # Correct page numbering
    print(f"Scraping Page: {page} → {url}")

    # ✅ Send GET request with UTF-8 encoding
    response = requests.get(url, headers=headers)

    # ✅ Ensure correct encoding
    response.encoding = "utf-8"

    # ✅ Check if request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # ✅ Find all news articles inside the 'category_content_listing' div
        articles = soup.find_all("div", class_="category_content_listing")

        for article in articles:
            # ✅ Extract Headline (inside <h3><a>)
            title_tag = article.find("h3").find("a") if article.find("h3") else None
            title = title_tag.get_text(strip=True) if title_tag else "No Title"

            # ✅ Extract Description (inside <p>)
            desc_tag = article.find("p")
            description = desc_tag.get_text(strip=True) if desc_tag else "No Description"

            # ✅ Extract Article Link
            link = "https://www.andhrajyothy.com" + title_tag["href"] if title_tag and "href" in title_tag.attrs else "No Link"

            # ✅ Append to list
            news_list.append({"Title": title, "Description": description, "Link": link})

        # ✅ Avoid Overloading Server (Respectful Scraping)
        time.sleep(2)  # Pause 2 seconds before next request

    else:
        print(f"❌ Failed to fetch Page {page}. HTTP Status: {response.status_code}")

# ✅ Convert to DataFrame and display
df = pd.DataFrame(news_list)

# ✅ Display DataFrame
from IPython.display import display
display(df)
print("✅ Scraping complete! Data saved.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# ✅ Base URL for pagination
base_url = "https://www.andhrajyothy.com/business/page/"

# ✅ Headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.90 Safari/537.36"
}

# ✅ Lists to store results
headlines = []
descriptions = []
links = []

# ✅ Loop through pages 2 to 12
for page in range(2, 21):  # Scraping pages 2 to 12
    url = f"{base_url}{page}"  # Correct pagination URL
    print(f"Scraping Page: {page} → {url}")

    # ✅ Send GET request with UTF-8 encoding
    response = requests.get(url, headers=headers)
    response.encoding = "utf-8"  # Fixes Telugu character encoding

    # ✅ Check if request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # ✅ Find all <h3> tags containing <a> links
        a_tags = soup.select("h3 > a")

        for a_tag in a_tags:
            # ✅ Extract Headline from <a> tag
            headline = a_tag.get_text(strip=True)

            # ✅ Find the next <p> tag after the <a> tag
            p_tag = a_tag.find_next("p")
            description = p_tag.get_text(strip=True) if p_tag else "No Description"

            # ✅ Extract Article Link
            link = "https://www.andhrajyothy.com" + a_tag["href"] if "href" in a_tag.attrs else "No Link"

            # ✅ Append to lists
            headlines.append(headline)
            descriptions.append(description)
            links.append(link)

        # ✅ Avoid Overloading Server (Respectful Scraping)
        time.sleep(2)  # Pause 2 seconds before next request

    else:
        print(f"❌ Failed to fetch Page {page}. HTTP Status: {response.status_code}")

# ✅ Convert to DataFrame
df = pd.DataFrame({"Headline": headlines, "Description": descriptions, "Link": links})

# ✅ Display DataFrame Correctly
from IPython.display import display
display(df)

print("✅ Scraping complete! Data saved.")
```

```python
df.to_csv('abn_telugu_business.csv', index=False)
print("DataFrame saved to abn_telugu_business.csv")
```
